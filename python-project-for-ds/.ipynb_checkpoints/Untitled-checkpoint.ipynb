For a brief outline, we want to create 4 objects for any given company in this stock market scenario:
- a yfinance Ticker
- a JSON file
- a Dictionary
- a Pandas Dataframe

### Setting up our Environment

```python
# first, we need to install our stock api yfinance and pandas

!pip install yfinance==0.2.4
!pip install wget

import yfinance as yf
import pandas as pd
import os.path
import wget
import json
```

### Creating our Objects - Ticker, JSON, Dict, and DataFrame

Using the `Ticker` module, we can create a Ticker object for a given company. We can then use this object's functions to call for data. For clarity on the yfinance library, read here: https://aroussi.com/post/python-yahoo-finance

```python
# instantiate a Ticker obj
apple = yf.Ticker('AAPL')
```

Now we can download the json file using the function `wget`. The function takes a `URL` as a parameter. I find it best to set a variable to this url string for readability.

```python
# clarify the url - for readability's sake
apple_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json'

# if the json file is not downloaded, download it using wget
if not os.path.isfile('apple.json'):
    wget.download(apple_url)
```

Finally, we can convert this JSON file to a `dict` using the `json.load` function. 

```python
# convert the json file to a python dict
with open('apple.json') as json_file:
        # use json to convert the file to a dict
        apple_info = json.load(json_file)
```

As a dictionary of key-value pairs, the apple_info object can be much easier to parse through. For example, if we wanted to know the `country` of a given company, we can simply search for the value at key `country`.


```python
apple_info['country']
```

#### Though we aren't done quite yet...

From here, we can also utilize the `history()` method given to us by yfinance to have a Pandas DataFrame of our stock history. Using the `period` parameter, we can also specify how far back we want our DataFrame to go.

Options for `period` inputs include `1d`, `1mo`, `1y`, `ytd`, and `max`, where 1 can be any integer value.

```python
apple_share_price_data = apple.history(period="20y")
```

The format that the data is returned in is a Pandas DataFrame. With the `Date` as the index the share `Open`, `High`, `Low`, `Close`, `Volume`, and `Stock Splits` are given for each day.
___________
To alter this we can reset the index of the DataFrame with the `reset_index` function. We can also set the `inplace` parameter to `True` so the change takes place to the DataFrame itself. That way, we can parse through the DataFrame with indeces rather than specific dates.


```python
apple_share_price_data.reset_index(inplace=True)
```

### All that's left is visualization

Now that we have our objects instantiated, we can use multiple different approaches to understand our data. Parsing through our dictonary may give us important descriptors for what kind of information we have available. We can also use Pandas functions to visualize our data and find hidden patterns.

```python
apple_share_price_data.head() # visualizing the first few rows
```

```python
apple_share_price_data.plot(x='Date', y='Open')
```
